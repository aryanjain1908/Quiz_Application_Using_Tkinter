```python
#importing required modules
import tkinter as tk
from tkinter import *
from tkinter import messagebox
from PIL import Image , ImageTk
import json
import time

#creating global variable which stores the option data
option = ""
checked = ""

# this function checks whether the option is correct or not
def check(start , name , age , data , credit , count):
    global checked
    global option
    if option == data["Answer"]:
        checked.configure(bg="green")
        credit+=100
        update_credit(name,age,credit)
        messagebox.showinfo("Information","Correct Answer")
        start_kbc(start,credit,str(int(count)+1),name,age)
    elif checked=="" or option=="":
        messagebox.showinfo("Information","Question is compulsary")
    else:
        checked.configure(bg="red")
        messagebox.showinfo("Information","Wrong Answer")
        account_page(start,name,age)
        
    
# this function update the credits
def update_credit(name,age,credit):
    with open("account.json" , "r") as f:
        data = json.load(f)
        
    for i in data:
        if i["Name"]==name and i["Age"]==age:
            i["Credits"] = credit
            
    with open("account.json","w") as f:
        json.dump(data,f)
        
# this function change color of button on clicking
def change_colour(start , a , b , c , d , change , op):
    global option 
    global checked
    option = change
    checked = op
    a.configure(bg="indigo")
    b.configure(bg="indigo")
    c.configure(bg="indigo")
    d.configure(bg="indigo")
    op.configure(bg="orange")

# this function displays questions and options   
def start_kbc(start , credit , count , name , age):
    global option
    global checked
    option = ""
    checked = ""
    
    
    if count=="11":
        account_page(start,name,age)
        return
    
    if count=="1":
        messagebox.showinfo("Information","This quiz contains 10 questions and each question has 100 credits")
    
    label = Label(start,bg="dark blue",width=600,height=400)
    label.place(x=0,y=0)
    
    with open("questions.json","r") as f:
        data = json.load(f)
        
    data = data[0]
    
    question = Label(start , text="Question "+count+" :" , justify = "center" , bg="dark blue" , fg = "white",font=("Ariel",20,"bold"),width=37)
    question.place(x=0,y=0)
    statement = Label(start , text=data[count]["Question"] , justify="center" , bg="dark blue" , fg = "white",font=("Ariel",15,"bold"),width=50)
    statement.place(x=0,y=30)
    
    a = tk.Button(start , command=lambda:change_colour(start ,a,b,c,d,"A",a) , text=data[count]["Options"]["A"] , justify="center" , bg="indigo" , fg="white" , font=("Ariel",15,"bold"),width=50)
    a.place(x=0,y=100)
    b = tk.Button(start , command=lambda:change_colour(start ,a,b,c,d,"B",b), text=data[count]["Options"]["B"] , justify="center" , bg="indigo" , fg="white" , font=("Ariel",15,"bold"),width=50)
    b.place(x=0,y=145)
    c = tk.Button(start , command=lambda:change_colour(start ,a,b,c,d,"C",c), text=data[count]["Options"]["C"] , justify="center" , bg="indigo" , fg="white" , font=("Ariel",15,"bold"),width=50)
    c.place(x=0,y=190)
    d = tk.Button(start , command=lambda:change_colour(start ,a,b,c,d,"D",d), text=data[count]["Options"]["D"] , justify="center" , bg="indigo" , fg="white" , font=("Ariel",15,"bold"),width=50)
    d.place(x=0,y=235)
    
    s = tk.Button(start , command=lambda:check(start , name , age , data[count] , credit , count), text="Submit" , justify="center" , bg="green" , fg="white" , font=("Ariel",15,"bold"),width=50)
    s.place(x=0,y=310)
    q = tk.Button(start , command=lambda:account_page(start,name,age), text="Quit" , justify="center" , bg="red" , fg="white" , font=("Ariel",15,"bold"),width=50)
    q.place(x=0,y=355)
    
# this function shows account page
def account_page(start , name , age):
    
    with open("account.json","r") as f:
        data = json.load(f)
      
    credits = 0
    for i in data:
        if i["Name"]==name and i["Age"]==age:
            credits = i["Credits"]
            
    label = Label(start , bg = "dark blue" , width=600 , height=400)
    label.place(x=0,y=0)
    label_name = Label(start , bg="dark blue" , text="Welcome! "+name , font=("Ariel" , 30 , "bold") , fg = "white",width=25 , justify="center")
    label_name.place(x=0 , y=50)
    label_credit = Label(start , bg="dark blue" , text="Credits : "+str(credits) , font=("Ariel" , 30 , "bold") , fg = "white" , width=25 , justify="center")
    label_credit.place(x=0,y=120)
    start_quiz = tk.Button(start , text="Start Quiz" , bg = "blue" , fg = "white" , font=("Ariel",30,"bold"),width=25 , command= lambda : start_kbc(start , credits , "1" ,name,age))
    start_quiz.place(x=0,y=330)
    
# this function manages account
def account(start , name , age):
    new_data = {"Name":name , "Age":age , "Credits":0}
    
    with open("account.json" , "r") as f:
        data = json.load(f)
        
    flag = False
    for i in data:
        if i["Name"]==name and i["Age"]==age:
            flag = True
            break
            
    
    
    if flag:
    
        pass
    
    else:
        
        
        data.append(new_data)


        with open("account.json" , "w") as f:
            json.dump(data,f)
        
    account_page(start , name , age)
    
# this function manages data
def data(start):
    label = Label(start , bg = "dark blue" , width=600 , height=400)
    label.place(x=0,y=0)
    name = Label(start , text="Enter your name" , font=("Ariel" , 20 , "bold") , bg="dark blue" , fg = "white")
    name.place(x=50,y=30)
    text_name = StringVar()
    take_name = Entry(start , bg="indigo" , fg="white"  , font=("Ariel" , 20 , "bold") , textvariable=text_name )
    take_name.place(x=50,y=70)
    age = Label(start , text="Enter your age" , font=("Ariel" , 20 , "bold") , bg="dark blue" , fg = "white")
    age.place(x=50,y=150)
    text_age = StringVar()
    take_age = Entry(start , bg="indigo" , fg="white"  , font=("Ariel" , 20 , "bold") , textvariable=text_age )
    take_age.place(x=50,y=190)
    button = tk.Button(start , text="Continue ->" , font=("Ariel" , 30 , "bold") , bg="blue" , fg="white" , width=25 , command = lambda : account(start , text_name.get() , text_age.get()))
    button.place(x=0,y=330)

# main function
def main():
    main_window = tk.Tk()
    main_window.title("KBC - Kaun Banega Krorepati")
    main_window.geometry(("600x400"))
    main_window.minsize(600 , 400)
    main_window.maxsize(600 , 400)
    image = Image.open("dummykbc.png")
    r_image = image.resize((600,400))
    n_image = ImageTk.PhotoImage(r_image)
    label = Label(main_window , image=n_image)
    label.place(x=0,y=0)
    start = tk.Button(main_window , text="Start" , fg="white" , font = ("Ariel" , 30 , "bold")  , width=25 , bg="blue"  , command = lambda : data(main_window))
    start.place(x=0,y=330)
    main_window.mainloop()

# calling main function
if __name__=="__main__":
    main()
```
